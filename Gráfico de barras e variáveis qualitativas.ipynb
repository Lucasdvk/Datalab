1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df = pd.read_csv('./previsao_de_renda.csv')



# Converter a coluna 'data_ref' para o tipo datetime
df['data_ref'] = pd.to_datetime(df['data_ref'])

# Gráfico de barras para Posse de Imóvel ao longo do tempo
plt.figure(figsize=(12, 6))
sns.countplot(x='data_ref', hue='posse_de_imovel', data=df)
plt.title('Distribuição da Posse de Imóvel ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.legend(title='Posse de Imóvel', loc='upper right')
plt.show()

# Gráfico de barras para Posse de Veículo ao longo do tempo
plt.figure(figsize=(12, 6))
sns.countplot(x='data_ref', hue='posse_de_veiculo', data=df)
plt.title('Distribuição da Posse de Veículo ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.legend(title='Posse de Veículo', loc='upper right')
plt.show()
````
